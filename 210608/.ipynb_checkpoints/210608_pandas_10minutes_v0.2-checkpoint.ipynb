# 6.Merge(데이터 합치기)

```python
df = pd.DataFrame(np.random.randn(10,4)); df
```

```python
peices = [df[:3]]; peices
```

```python
peices = [df[:3], df[3:7]]; peices
```

```python
peices = [df[:3], df[3:7], df[7:]]; peices
```

```python
pd.concat(peices)
```

```python
# Join(결합) 
left = pd.DataFrame({'key':['foo','foo'],
                    'lval':[1,2]}) ; print("left = ", left)
right = pd.DataFrame({'key':['foo','foo'],
                     'rval':[4,5]}); print("right = ", right)
```

```python
pd.merge(left, right, how='inner', on = 'key')
```

```python
pd.merge(left, right, how='left', on = 'key')
```

```python
pd.merge(left, right, how='right', on = 'key')
```

```python
pd.merge(left, right, how='outer', on = 'key')
```

```python
how : {'left', 'right', 'outer', 'inner'}, default 'inner'
```

```python
left = pd.DataFrame({'key':['foo','bar'],
                    'lval':[1,2]}) ; print("left = ", left)
right = pd.DataFrame({'key':['foo','bar'],
                     'rval':[4,5]}); print("right = ", right)
```

```python
pd.merge(left, right, on = 'key')
```

# 7.Grouping(그룹화)

### -.결과물들을 하나의 데이터 구조로 결합(combining) 

```python
df = pd.DataFrame({'A':['foo','bar','foo','bar','foo','bar','foo','foo'],
                  'B':['one','one','two','three','two','two','one','three'],
                  'C':np.random.randn(8),
                  'D':np.random.randn(8)});df
```

```python
# 'A'기준으로 그룹화 후 각 그룹에 sum() 적용 
df.groupby('A')
```

# 10.Categoricals(범주화)

```python
# 범주형 데이터 

df = pd.DataFrame({'id':[1,2,3,4,5,6], 
                  'raw_grade':['a','b','b','a','a','e']});df
```

```python
df['grade'] = df['raw_grade'].astype('category')
```

```python
df['grade']
```

```python
# 범주에 이름 붙이기
df["grade"].cat.categories = ["very good","good","very bad"];df
```

```python
# 범주의 순서를 바꾸고, 동시에 누락된 범주 추가
df["grade"] = df["grade"].cat.set_categories(["vary bad","bad",
                                              "medium","good",
                                              "very good"])
```

```python
df["grade"]
```

```python
df.sort_values(by="grade")
```

```python
df.groupby("grade").size()
```
