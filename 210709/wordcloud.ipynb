```python
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import pandas as pd
```

```python
wc = WordCloud('/content/drive/MyDrive/Colab Notebooks/font/Typo_SsangmunDongB.ttf')

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/신혼부부_naverkin.csv')
df
```

```python
text = ''
for i in range(len(df)) :
  text += df.loc[i, 'title']
  text += str(df.loc[i, 'content'])
  text += df.loc[i, 'answer']
```

```python
okt = Okt()
noun = okt.nouns(text)
for i, v in enumerate(noun) :
  if len(v) < 2 :
    noun.pop(i)

count = Counter(noun)
noun_list = count.most_common(100)
noun_list
```

```python
keyword = '신혼부부'
wc.generate_from_frequencies(dict(noun_list))
wc.to_file('/content/drive/MyDrive/Colab Notebooks/data/' + keyword + '_naverkin.png')
```
