```python
# 결측치 (Missing Data)
# CSV 파일 읽어오기

import pandas as pd

path = "../bicycle/bicycle_dup.csv"

df = pd.read_csv(path, encoding='cp949')

df
```

```python
# 2. 데이터 확인
# 이용시간과 이용거리가 상관관계가 있다는 가정하에 이용시간이 결측치이면 발생된 결측치의 
# 이용거리와 같은 이용거리의 다른 데이터들에 대한 평균 이용시간을 확인해 보려고 한다.
```

```python
# 이용시간이 결측치인 모든 데이터 추출

df[df['이용시간'].isnull()]
```

```python
# 이용거리가 2290 인 데이터만 출력

df[df['이용거리']==2290]
```

```python
# 이용거리를 기준으로 그룹화 시키고 그것에 따른 평균값을 구했다. 

df_group = df.groupby('이용거리')
df_group.mean()
```

```python
print(df_group.mean().loc[2290])
```

```python
df_group.mean().loc[2290]['이용시간']
```

```python
# 데이터 중복확인
# Dataframe에서는 중복 확인을 위해 duplicated라는 함수를 지원한다. 
# 중복된 데이터가 나타나면 true를 반환한다.
```

```python
print(df['이용거리'][:5])
print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
print(df.duplicated(['이용거리'][:5]))
```

```python
print(df)
print(df.duplicated(['이용거리']))
df.duplicated(['이용거리', '나이'])
```

```python
df.duplicated(['이용거리'], keep='last')
```

```python
print( df.duplicated(['이용거리']).sum() )
print( df.duplicated(['이용거리'], keep='last').sum() )
print( df.duplicated(['이용거리'], keep=False).sum() )
df.duplicated(['이용거리'], keep=False)
```

```python
# 중복데이터 제거

df.drop_duplicates(['이용거리'], keep=False)
```
