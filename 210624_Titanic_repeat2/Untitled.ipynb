```python
import pandas as pd

path = '../data/Titanic/titanic.csv'

df = pd.read_csv(path, encoding='cp949')
```

```python
df
```

```python
df.isnull()
```

```python
import seaborn as sns

sns.heatmap(df.isnull())
```

```python
# heatmap : heat 는 온도라는 뜻으로.. 두 객체? 를 비교할때 쓴다고는 들었다...

# heatmap 은 값의 변화를 온도로 표현하여 변화량을 확인 할수 있는 그래프
# 타이타닉 정보의 결측치 유무(isnull)를 전달하여 결측치가 나타나는 부분을 True(1)로 
# 아닌 부분은 False(0)으로 표현하게 된다.
```

```python
sns.heatmap(df.isnull(), center=0.4)
```

```python
sns.heatmap(df.isnull(), center=0.4 , linewidths=0.001)
```
