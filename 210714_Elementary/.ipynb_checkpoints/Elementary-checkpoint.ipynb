```python
import pandas as pd

path = '../data/Seoul_school/Seoul_school.csv'

df = pd.read_csv(path, encoding='cp949')

df
```

```python
for col in df.columns:
    print(col)
```

```python
df_school = df[['학교명','도로명주소','도로명상세주소']]
```

```python
df_school
```

```python
df_school.rename(columns={'도로명주소':'구명'}, inplace=True)
df_school.rename(columns={'도로명상세주소':'동명'}, inplace=True)
```

```python
df_school
```

```python
df_gu = df_school['구명'].str.findall('\w+구')
```

```python
df_gu
```

```python
df_school['구명'] = df_gu
```

```python
df_Dong = df_school['동명'].str.findall('\w+동')
```

```python
df_Dong
```

```python
df_school['동명'] = df_Dong
```

```python
df_school
```

```python
df_element = df_school.loc[df_school['학교명'].str.contains("초등학교")]
```

```python
df_element
```

```python
df_element.rename(columns={'학교명':'초등학교명'}, inplace=True)
```
