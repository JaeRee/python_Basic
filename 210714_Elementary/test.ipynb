```python
import pandas as pd

path = 'element2.csv'

df = pd.read_csv(path, encoding='cp949')

df
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df2=df.groupby('구명').count()
```

```python
df2
```

```python
df2.rename(columns={'초등학교명':'초등학교 갯수'}, inplace=True)
```

```python
df2.drop(['동명'], axis=1, inplace=True)
```

```python
df2
```

```python
# CSV 파일로 저장하기

df2.to_csv('element_count.csv', mode='a', header=True, encoding='cp949')

# 
```
