````python
import pandas as pd

df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
````

````python
# 1. 도시별 가격 평균을 구하시오

df_city = df.groupby('city')
df_city.mean()
````

````python
# 2. 각 도시에서 과일별 가격 평균을 구하시오

df_city.describe()
````

````python
df_city.get_group('서울')
````

````python
df_seoul = df_city.get_group('서울')
df_seoul.groupby('fruits')
````

````python
df_2 = df_seoul.groupby('fruits')
df_2.describe()
````

````python
df_city.get_group('부산')
````

````python
df_busan = df_city.get_group('부산')
df_busan.groupby('fruits').describe()
````

````python
# 2. 각 도시에서 과일별 가격 평균을 구하시오

df_mean = df.groupby(['city', 'fruits'])

df_mean.mean()
````
